```python
pivotdf = df.pivot_table(index='genre', columns='topic', aggfunc='size')
```

```python
# normalize
normalizeddf = pivotdf.div(pivotdf.sum(axis=1), axis=0)

normalizeddf.plot(kind='bar', stacked=False, figsize=(12,6))
plt.title("Normalized Distribution of Topics Across Genres")
plt.xlabel("Genre")
plt.ylabel("Proportion of Songs")
plt.legend(title="Topic", bbox_to_anchor=(1, 1))
plt.show()
```

```python
normalizeddf
```

songs in each genre associated with different lyrical topics
- hip hop has a big proportion obscene content (about .77) 
- country has the highest sadness score (about .29)
- rock has the strongest presence of violence (about .32)
- world/life themes are most prominent in reggae (about .24)
- pop, blues, and jazz have broad thematic range and dominance in the dataset
- sadness, world/life, violence are most prominent (obscense follows close)

normalizing the data because genres have varying numbers of songs in the dataset; comparing relative proportions rather than raw counts, highlights the topic preferences unique to each genre.

Users who enjoy:
- hip hop with less obscene content might like blues or jazz 
- romantic themes, might like country and pop suggestions
- rock fans interested in less violent content might like pop or country


```python
genretopic = pd.crosstab(df['genre'], df['topic'], normalize='index')
plt.figure(figsize=(14, 8))
sns.heatmap(genretopic, annot=True)
plt.title("Topics Across Genres")
plt.xlabel("Topic")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

genre specific findings are the same as the above, as for cross genre patterns its similar but a bit more clear:
- feelings is consistently low across all genres (darkest column)
- violence, sadness, and world/life are most prominent 
- obscene is bright across genres as well, but more dramatic differences

```python
lyricscores = ['dating', 'violence', 'world/life', 'night/time', 'shake the audience','family/gospel', 'romantic', 'communication', 'obscene', 'music','movement/places', 'light/visual perceptions', 'family/spiritual','sadness', 'feelings']
```

```python
topicscores = df.groupby('topic')[lyricscores].mean()

plt.figure(figsize=(14, 8))
sns.heatmap(topicscores, annot=True, fmt=".2f")
plt.title("Average Lyric Scores by Topic")
plt.xlabel("Lyric Score")
plt.ylabel("Topic")
plt.tight_layout()
plt.show()
```

lyrical scores relate to the topic label 
- violence topic high correlation with violence lyrical score (.43)
- world/life shows high correlation with itself (.43)
- romantic shows high correlation with itself (.42)
- obscene shows high correlation with itself (.46)
- sadness shows high correlation with itself (.43) 
- (and so on)

stong correlation with self-reinforcing topics and lyric classfications

```python
genrescores = df.groupby('genre')[lyricscores].mean()

plt.figure(figsize=(14, 8))
sns.heatmap(genrescores, annot=True, fmt=".2f")
plt.title("Average Lyric Scores by Genre")
plt.xlabel("Lyric Score")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

genre relates with lyrical score
- hip hop strong association with obscene content (0.41)
- rock has highest scores for violence (0.18)
- country and blues show stronger associations with sadness (0.17 and 0.11)
- reggae has a strong presence of violence, obscence and world/life (0.14)
- family/gospel content is consistently low across all genres

(simialr findings to genre and topic heatmap)

heatmaps help confirm that topic and genre strongly correlate with certain lyrical features
